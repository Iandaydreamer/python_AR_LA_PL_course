```python
def fibonacci(s,count): 
    fib_lst=[0,1]
    fib_part=[]
    if s==0 or s==1:
        fib_part[:]=fib_lst
        for i in range(count-2):
            fib_part.append(fib_part[-1]+fib_part[-2])
    else:
        while True:
            fib_lst[:]=[fib_lst[1],fib_lst[0]+fib_lst[1]]
            #print(fib_lst)
            if fib_lst[1]-s>=0:break
        fib_part[:]=fib_lst
        if abs(fib_lst[0]-s)>=abs(fib_lst[1]-s):
            for i in range(count-1):
                fib_part.append(fib_part[-1]+fib_part[-2])
            fib_part.pop(0)
        else:
            for i in range(count-2):
                fib_part.append(fib_part[-1]+fib_part[-2])
    return fib_part

print("fibonacci(6,8)->{}".format(fibonacci(6,8)))
print("fibonacci(4,7)->{}".format(fibonacci(4,7)))
```

```python
class Bird:
    fly='Whirring' 
    def __init__(self):
        self.hungry=True
    def eat(self):
        if self.hungry:
            print('Aaaah...')
            self.hungry=False
        else:
            print('No.Thanks!')

class Apodidae(Bird):  
    def __init__(self):
        super(Apodidae,self).__init__()
        self.sound='Squawk!'  
    def sing(self):
        print(self.sound)
swift=Apodidae()
print("swift.fly->{}".format(swift.fly))
print("swift.eat()->")
swift.eat()
print("swift.eat()->")
swift.eat()
print("swift.sing()->")
swift.sing()
```

```python
lst=[list(range(2,15,2)),[5,8,77,list(range(2,9,3)),6]]
print("lst={}".format(lst))
flatten_lst=[]  #返回一个一维数组
for v_1 in lst:  #v代表value
    try:
        for v_2 in v_1:  #循环列表 
            try:
                for v_3 in v_2:     
                    flatten_lst.append(v_3)
            except TypeError:
                flatten_lst.append(v_2)
    except TypeError:   #当函数被告知展开一个元素时，引发TypeError异常，生成器返回一个值
        flatten_lst.append(v_1)
print("flatten_lst={}".format(flatten_lst))
```

```python
def f_convert_a(x):
    try:
        return float(x)  #float()函数返回一个十进制浮点型数值（小数）里面包含三种类型数据1.二进制、八进制、十进制、十六进制的整数。 2.bool（布尔值True和False）。 3.表示十进制数字的字符串。
    except:
        return x
print("f_convert_a('3.1415')->{}".format(f_convert_a('3.1415')))    
print("f_convert_a('string')->{}".format(f_convert_a('string')))  
print("f_convert_a(7)->{}".format(f_convert_a((7))))  
```

```python
def f_convert_b(x):
    try:
        return float(x)
    except ValueError:   #意味着except所捕获的异常行为
        return x
print("f_convert_b('7')->{}".format(f_convert_b('7')))    
print("f_convert_b('string')->{}".format(f_convert_b('string')))  
print("f_convert_b(5,4,3)->{}".format(f_convert_b((5,4,3))))  
```

```python
def f_convert_c(x):
    try:
        return float(x)
    except ValueError:
        return x
    except TypeError:
        print(x,'TypeError')
print("f_convert_c(7')->{}".format(f_convert_c('7')))    
print("f_convert_c('string')->{}".format(f_convert_c('string')))  
print("f_convert_c(5,4,3)->{}".format(f_convert_c((5,4,3))))         
```

```python
def f_convert_f(x):
    try:
        return float(x)
    except (ValueError,TypeError) as e: 
          pass #使用pass语句可以忽略异常
print("f_convert_f('3.1415')->{}".format(f_convert_f('3.1415')))    
print("f_convert_f('string')->{}".format(f_convert_f('string')))  
print("f_convert_f(3,6,7)->{}".format(f_convert_f((3,6,7))))     
```

```python
def f_open_a(fp):
    try:
        f=open(fp,'r')
    except IOError as e:
        print('Unable to open',fp,':%s\n' %e)
    else:  #如果try代码块中的程序没有引发异常，将会执行else子句中的程序
        data=f.read()  #逐个字节或者字符读取文件中的内容
        f.close
        return data
tryException_content=f_open_a("./data/tryException.txt")   
print("tryException_content->{}".format(tryException_content))
```

```python
def f_open_b(fp):
    try:
        f=open(fp,'r')
    except IOError as e:
        print('Unable to open',fp,':%s\n' %e)
    else:
        data=f.read()
        f.close()
        return data
    finally:  #finally子句提供一些程序代码，无论程序是否出错误都需要执行该代码
        print('done!')
f_open_b("./data/tryException.txt")        
```
